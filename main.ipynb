```python
df['mort_acc'].value_counts()
```

```python
#it might be hard to drop the mortgage account column since it consists of a large part of our data

df.corr()['mort_acc'].sort_values()
```

We can see that total_acc is the most corrlelated to mort_acc

```python
mort_acc_avg = df.groupby('total_acc').mean()['mort_acc']
```

```python
def replace_mort(total_acc, mort_acc):


    if np.isnan(mort_acc):
        return mort_acc_avg[total_acc]
    else:
        return mort_acc
```

```python
df['mort_acc'] = df.apply(lambda x: replace_mort(x['total_acc'], x['mort_acc']), axis=1)
```

```python
df.isnull().sum()
```

```python
#there are still null values, but its a super small fraction of our entire data set.

df = df.dropna()
```

```python
df.isnull().sum()
```

Now we have no more null values!

```python
df['term'].value_counts()
```

```python
#instead of getting 36 months and 60 columns, we can do 36 and 60 months

df['term'] = df['term'].apply(lambda term: int(term[:3]))
```

```python
df = df.drop('grade', axis=1)
```

```python
df['sub_grade']
```

```python
dummies = pd.get_dummies(df['sub_grade'], drop_first=True)

df = pd.concat([df.drop('sub_grade', axis=1),dummies],axis=1)
```

```python
df.columns
```

```python
dummies = pd.get_dummies(df[['verification_status', 'application_type','initial_list_status','purpose']], drop_first=True)

df = pd.concat([df.drop(['verification_status', 'application_type','initial_list_status','purpose'], axis=1),dummies],axis=1)
```

```python
df.columns
```

Turning all the types of home ownership into dunny variables so the model can numerically interpret the data

```python
df['home_ownership'].value_counts()
```

```python
df['home_ownership'] = df['home_ownership'].replace(['NONE', 'ANY'], 'OTHER')
```

```python
house_dummies = pd.get_dummies(df['home_ownership'], drop_first=True)
df = df.drop('home_ownership', axis=1)
df = pd.concat([df,house_dummies], axis=1)
```

```python
# the date where the loan was issued isn't important
df = df.drop('issue_d', axis=1)
```

```python
#created a column based on the zipcode extracted from the address column

df['zip_code'] = df['address'].apply(lambda address:address[-5:])
```

```python
dummies = pd.get_dummies(df['zip_code'],drop_first=True)
df = df.drop(['zip_code','address'],axis=1)
df = pd.concat([df,dummies],axis=1)
```

```python
df['earliest_cr_year'] = df['earliest_cr_line'].apply(lambda date:int(date[-4:]))
df = df.drop('earliest_cr_line',axis=1)
```

```python
dummies = pd.get_dummies(df['loan_status'],drop_first=True)
df = df.drop('loan_status', axis=1)
df = pd.concat([df,dummies], axis=1)
```

Splitting the DataSet


```python
from sklearn.model_selection import train_test_split
```

```python
df.head()
```

```python
X = df.drop('Fully Paid',axis=1).values
y = df['Fully Paid'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=101)
```

Normalizing the Data


```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

Model Creation and Training


```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train, y_train)
```

```python
predictions = rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
predictions = rfc.predict(X_test)
```

```python
print(classification_report(y_test,predictions))
```
